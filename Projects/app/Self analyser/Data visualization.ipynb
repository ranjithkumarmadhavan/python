```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
dataFrame = pd.read_csv("tables/time_sheet_master.csv")
```

```python
dataFrame
```

```python
## Trend of work , sleep and learning
```

```python
x = dataFrame["date"]
work = dataFrame.work.fillna(dataFrame.work.mean())
learning = dataFrame.learning.fillna(dataFrame.learning.mean())
plt.plot(x,dataFrame['sleep'],label='sleep',color='red')
plt.plot(x,work,label='work',color='blue')
plt.plot(x,learning,label='learning',color='yellow')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```
